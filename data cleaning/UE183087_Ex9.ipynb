### Question 2. From all marks based attributes, remove % sign from different attributes. In SAT/ACT, Keep higher value.

```python
#defining Sat converter
def satconverter(x):
    # removing % from  *SAT or ACT
    x = str(x)
    if not '/' in x:
         return np.nan
    else:
        return max(float(x.split('/')[0].replace("*","")),float(x.split('/')[1].replace("%", "").replace("*","")))
```

```python
#printing the data
data["*SAT or ACT"]= data["*SAT or ACT"].apply(satconverter)
data.head(25)
```

### Question 3. From all cost based attributes, Remove $ sign and commas.

```python

comma = ','
dollar = '$'


def clean_costs(item):
    set= str(item.loc['Total Costs'])
    
 
    if comma in set:
        comma_index = set.find(comma)
        set = set[:comma_index] + set[comma_index + 1:]
    
    return set

    if set == 'nan':
        return np.NaN
    
    set = set[1:]
   

data['Total Costs'] = data.apply(clean_costs, axis = 1)
```
